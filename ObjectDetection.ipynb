### Import IMAGEAI object detection libraries

```python
from imageai.Detection import ObjectDetection
import os
import cv2
import matplotlib.pyplot as plt
import time
from PIL import Image
from tensorflow.python.framework.ops import disable_eager_execution
disable_eager_execution()
```

### Create YOLO model for object detection in an image

```python

yolo_obj = ObjectDetection()
yolo_obj.setModelTypeAsYOLOv3()
yolo_obj.setModelPath("/content/yolo.h5")
yolo_obj.loadModel()
```

### Implement the object detection function in an image

### Display detected object from image

```python
detections = yolo_obj.detectObjectsFromImage(input_image = ("/image.jpg"), output_image_path = ("/output_image.jpg"))

for objects in detections:
    print(objects["name"] , " : " , objects["percentage_probability"] )

Image.open("/output_image.jpg")
```

```python
detections = yolo_obj.detectObjectsFromImage(input_image = ("/child.jpg"), output_image_path = ("/childimage_output.jpg"))

for objects in detections:
    print(objects["name"] , " : " , objects["percentage_probability"] )

Image.open("/childimage_output.jpg")
```
